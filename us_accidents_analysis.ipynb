## Start Time (column exploring)

```python
df.Start_Time
```

```python
df.Start_Time[0]
```

Now this is currently a string we convert into a DATE !!

- To parse the date we use (to_datetime) keyword

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
# df.Start_Time.dt.hour

sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

**Observation** - You can see in the histogram that is most of the accidents happend in 10 to 5 pm timing

- A high percentage of accidents occured in between 7am to 12am (Probably people in hurry to get to work
- Next highest percentage is 1pm to 6pm

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

**Obeservation** - most of the accidents occured in weekdays

Is the distrubution of accidents by hour the same on weekends as on weekdays !!

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
# sundays_start_time
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

## On sundays there is more spreadout distrubution 
- Most of the accidents occured in afternoon

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
# sundays_start_time
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

**Same here also most accidents occured in Afternoon only**.. 

By this you can check all day's analysis

```python
sns.distplot(monday_start_time.dt.month, bins=12, kde=False, norm_hist=True)
```

**Observation** - You can see that highest percentage accidents of **MONDAY** has the **AUGUST** to **DECEMBER**..
